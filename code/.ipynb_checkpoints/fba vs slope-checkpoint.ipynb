```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fba = pd.read_csv('../result/fba_abs.csv', index_col=0)
t = pd.read_csv('../result/slope.csv', index_col=0).T
v = pd.DataFrame(pd.read_excel('../result/lac.xlsx').mean()).T
v.columns = t.columns
slope = pd.concat([t, v])
slope = slope.rename(index={0:'Lactate'})
slope = pd.concat([slope, pd.DataFrame([22, 12.5, 17], columns=['Doubling Time'] , index=slope.columns).T])
```

```python
Lac = np.array(slope.T['Lactate']/slope.T.mean()['Lactate'])
dt = np.array([np.log(2)/i for i in slope.T['Doubling Time']])
DT = dt/dt.mean()
```

```python
slope.T['Lactate'].plot(kind='bar')
pd.Series(fba.loc['Lactate_output']/fba.loc['Lactate_output'][0]).plot()
```

```python
pd.Series(Lac/Lac[0]).plot()
pd.Series(fba.loc['Lactate_output']/fba.loc['Lactate_output'][0]).plot(kind='bar')
```

```python
pd.Series(DT).plot(kind='bar')
pd.Series(fba.loc['BIOMASS']/fba.loc['BIOMASS'][0]).plot()
```

```python
pd.Series(DT/DT[0]).plot()
pd.Series(fba.loc['BIOMASS']/fba.loc['BIOMASS'][0]).plot(kind='bar')
```
