```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
fba = pd.read_csv('../result/fba_abs.csv', index_col=0)
t = pd.read_csv('../result/slope.csv', index_col=0).T
v = pd.DataFrame(pd.read_excel('../result/lac.xlsx').mean()).T
v.columns = t.columns
slope = pd.concat([t, v])
slope = slope.rename(index={0:'Lactate'})
slope = pd.concat([slope, pd.DataFrame([22, 12.5, 17], columns=['Doubling Time'] , index=slope.columns).T])
```

```python
fig, ax = plt.subplots()

ax2 = ax.twinx()
p1 = ax.bar(['Con', 'Clone A', 'Clone B'], slope.T['Lactate'], color='black')
ax.set_ylim(0, 5)
ax.set_ylabel('Metabolome')
ax2.plot(['Con', 'Clone A', 'Clone B'], np.array(fba.loc['Lactate_output']), color='red')
ax2.set_ylabel('Flux')
ax.set_title('Lactate')
plt.show()
```

```python
fig, ax = plt.subplots()

ax2 = ax.twinx()
p1 = ax.bar(['Con', 'Clone A', 'Clone B'], [np.log(2)/i for i in slope.T['Doubling Time']], color='orange')
ax.set_ylim(0, 0.1)
ax.set_ylabel('Metabolome')
ax2.plot(['Con', 'Clone A', 'Clone B'], np.array(fba.loc['BIOMASS']), color='green')
ax2.set_ylabel('Flux')
ax.set_title('Biomass')
plt.show()
```
