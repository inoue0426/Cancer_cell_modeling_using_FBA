```python
import pandas as pd
import numpy as np
import urllib.request
import re
```

```python
t = pd.read_csv('../csv/network.csv', index_col=0)
```

```python
c = set(t['RNum'])
x = set(t['CNum'])
```

```python
v = [
    'oxalosuccinate carboxy-lyase',
    '(S)-malate:NAD+ oxidoreductase',
    'acetyl-CoA:oxaloacetate C-acetyltransferase',
    '(S)-malate:quinone oxidoreductase',
    '2-phospho-D-glycerate hydro-lyase',
    'D-glyceraldehyde-3-phosphate aldose-ketose-isomerase',
    'D-glyceraldehyde-3-phosphate:NAD+ oxidoreductase',
    'beta-D-fructose-1,6-bisphosphate D-glyceraldehyde-3-phosphate-lyase',
    'pyruvate:ferredoxin 2-oxidoreductase',
    '2-oxoglutarate:ferredoxin oxidoreductase',
    'citrate hydro-lyase',
    'ATP:3-phospho-D-glycerate 1-phosphotransferase',
    'D-phosphoglycerate 2,3-phosphomutase',
    'ATP:alpha-D-glucose 6-phosphotransferase',
    'Isocitrate:NADP+ oxidoreductase',
    'isocitrate hydro-lyase',
    'alpha-D-Glucose 6-phosphate ketol-isomerase',
    'ATP:D-fructose-6-phosphate 1-phosphotransferase',
    'succinyl-CoA:acetate CoA-transferase'
]
```

```python
d = pd.concat([
    pd.DataFrame(v), 
    pd.DataFrame(list(c))
], axis=1)
d.columns = ['Reaction','RNum']
```

```python
n = [
    'ATP',
    'NAD',
    'NADH',
    'NADPH',
    'NADP',
    'ADP',
    'CoA',
    'CO2',
    'Pyruvate',
    'Acetyl-CoA',
    '2-Oxoglutarate',
    'Oxaloacetate',
    'Succinate',
    'Phosphoenolpyruvate',
    'Succinyl-CoA',
    'Glycerone phosphate',
    'D-Glyceraldehyde 3-phosphate',
    'Reduced ferredoxin',
    'Oxidized ferredoxin',
    'Malic acid',
    'Citrate',
    '3-Phospho-D-glycerate',
    '3-Phospho-D-glyceroyl phosphate',
    'alpha-D-Glucose',
    'Isocitrate',
    'cis-Aconitate',
    '2-Phospho-D-glycerate',
    'alpha-D-Glucose 6-phosphate',
    'beta-D-Fructose 6-phosphate',
    'beta-D-Fructose 1,6-bisphosphate',
    'Oxalosuccinate'
]
```

```python
f = pd.concat([
    pd.DataFrame(n), 
    pd.DataFrame(list(x))
], axis=1)
f.columns = ['Compounds','CNum']
```

```python
t = t.sort_values('CNum').reset_index(drop=True)
f = f.sort_values('CNum').reset_index(drop=True)
```

```python
b = pd.merge(t, f, on='CNum', how='inner')
```

```python
b = pd.merge(b, d, on='RNum', how='inner')
```
