```r
## doubling time (h)
d2  <- 20.3

## growth rate (/day)
ll <- log(2) / d2 * 24

## model function
## y = C - A exp(lambda t)
y_pred <- function(ppp){
    return(ppp[1] - ppp[2]*exp(ll*t$day))
}

result <- optim(c(c0=3.0, aa=1),function(ppp){mean((t$y-y_pred(ppp))^2)})

plot(t,type="b")
points(t[,1], y_pred(result$par),type="l",col=2)
```
